##CONTEXTO

El siguiente dataset recopila información para 193 países entre los años 2000 y 2015. Hace foco en las condiciones de inmunización para enfermedades tales como Hepatitis B y Polio. Sumado a estos factores, también se muestran datos sobre la condición económica y social de cada país.

En otras palabras, este estudio recopila información de factores de inmunización, factores de mortalidad, factores económicos, factores sociales y otros factores relacionados con la salud, los cuales se podrán utilizar para poder entender en qué aspectos un país debe hacer foco a la hora de mejorar su expectativa de vida.

**Contenido**

El dataset es provisto por WHO (World Health Organization)

##ENCODING

Como parte de la práctica de OHE, transformamos los estados de Developed y Developing a 0 y 1.

```python
#ONE HOT ENCODING con GetDummies- Clase 40
from sklearn.preprocessing import OneHotEncoder
OHE=['Status']
ohe= OneHotEncoder()
df_OHE=pd.get_dummies(df_life[OHE], columns = OHE)
```

```python
df_OHE.head()
```

```python
df_life['Country'].value_counts()
```

##FEATURE TRANSFORMATION

Para los países que tenemos poca info, cambiamos su denominación a "Otro", ya que no son representativos.

```python
#FEATURE TRANSFORMATION
# Reconvertir categorias para hacerlas reducidas y disminuir ruido
df_life_tr = df_life
df_life_tr['Country']=[x if x in df_life_tr['Country'].value_counts()[0:1].index else 'Otro' for x in df_life_tr['Country']]

df_life_tr
```

##ML SUPERVISADO - Clasificación

Usaremos ML para estimar el campo Status, en función de los demás campos del dataset.

```python
#MODELO SUPERVISADO
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df_life_tr.isna().sum()
```

```python
df_life_ok = df_life_tr.dropna()
df_life_ok.isna().sum()
```

```python
X = df_life_ok.drop(columns=['Status','Country'])
y = df_life_ok['Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

###KNN

```python
neigh = KNeighborsClassifier(n_neighbors=15)
neigh.fit(X_train, y_train)
```

```python
y_pred=neigh.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred=y_pred, y_true= y_test))
```

###Random Forest y LOOCV

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
```

```python
# Crear el procedimiento loocv 
cv = LeaveOneOut()
```

```python
# Creamos modelo y hacemos FIT
clf = RandomForestClassifier(max_depth=6, random_state=42, n_estimators=70,criterion='gini')
clf.fit(X_train, y_train)
```

```python
# Evaluamos modelo según LOOCV (clase 48)
scores = cross_val_score(clf, X_train, y_train, scoring='accuracy', cv=cv, verbose=1)
```

```python
# Creamos reporte de Perfomance
from numpy import mean
from numpy import std
print('Accuracy del Modelo: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
#Probamos el Modelo
y_pred_clf = clf.predict(X_test)
```

```python
#Otro tipo de reporte para Accuracy del Modelo
print(classification_report(y_pred=y_pred_clf, y_true= y_test))
```

Se puede observar un mejor resultado de validación con el método de Random Forest.

Accuracy según Sklearn Metrics

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_clf)
```

```python
#Matriz de Confusión (se repite más abajo para separar la parte de métricas)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_clf) 
```
